```python
import numpy as np
import matplotlib.pyplot as plt

soa = np.array([[0, 0, 1, 2], [0, 0, 5, 1]])
X, Y, U, V = zip(*soa)
plt.figure()
ax = plt.gca()
ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)

ax.set_xlim([0, 6])
ax.set_ylim([0, 5])
plt.draw()
plt.show()
```

```python
import math 
vectors = np.array([[0, 0, 0, 0]])
vectors2 = np.array([[0, 0, 0, 0]])
startpoint=np.array([[0, 0]])
A=np.array([[1,2],[-1,1]])
#print(A)
for i in range(0,20):
    endpoint=np.array([[math.cos(18*i*math.pi/180),math.sin(18*i*math.pi/180)]])
    y=np.transpose(np.dot(A,np.transpose(endpoint)))
    #print(y)
    newvector = np.column_stack((startpoint,endpoint))
    newvector2 = np.column_stack((startpoint,y))
    #print(newvector)
    vectors=np.row_stack((vectors,newvector))
    vectors2=np.row_stack((vectors2,newvector2))
#print(vectors)
X, Y, U, V = zip(*vectors)
plt.figure()
ax = plt.gca()
ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)

ax.set_xlim([-3, 3])
ax.set_ylim([-2, 2])
plt.draw()
plt.show()

X, Y, U, V = zip(*vectors2)
plt.figure()
ax = plt.gca()
ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)

ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
plt.draw()
plt.show()
```
