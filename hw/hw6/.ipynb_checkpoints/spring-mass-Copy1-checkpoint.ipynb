```python
import numpy as np
import matplotlib.pyplot as plt
import math
import numpy
list = numpy.loadtxt(open("test3data.csv","rb"),delimiter=",",skiprows=1)
X=list[:,[1,2]].T
#print(X)
nPoints=X.shape[1]
plt.plot(X[0,:],X[1,:],'o')
#plt.xlim(xmin = 0, xmax = 200)
#plt.ylim(ymin = 0, ymax = 350)
plt.show()
```

```python
Xavg=np.mean(X, axis=1)
np.transpose(Xavg)
```

```python
B=X - np.diag(Xavg).dot(np.ones([2, nPoints]))
#B=B/math.sqrt(nPoints)
plt.plot(B[0,:],B[1,:],'o')
plt.show()
```

```python
from scipy.linalg import svd
#print(B/math.sqrt(nPoints))
# SVD
U, s, VT = svd(B,full_matrices=False)
print(U)
print(np.diag(s))
print(VT)

#s = s.reshape(s.shape[0],1)
#print(np.diag(s))
#print(U.dot(np.diag(s)))
```

```python
Y=U.T.dot(B)
print(U.T)
print(np.cov(Y))
print(np.cov(B))
#print(Y)
```
